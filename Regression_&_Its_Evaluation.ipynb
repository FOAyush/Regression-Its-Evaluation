# **1) What is Simple Linear Regression? **

**Ans)-** **Simple Linear Regression** is a fundamental statistical technique used to examine the relationship between two continuous variables. It models how a dependent variable (often called Y) changes as a function of an independent variable (X). The goal is to fit a straight line through the data points that best represents this relationship. This line is described by the equation $Y = \beta_0 + \beta_1 X + \varepsilon$, where $\beta_0$ is the intercept, $\beta_1$ is the slope of the line, and $\varepsilon$ represents the error term, accounting for variability not explained by the model. Simple linear regression assumes that the relationship between X and Y is linear, meaning changes in X produce proportional changes in Y. It is commonly used for predictive modeling—for example, predicting a student's exam score based on hours studied or forecasting sales based on advertising spending. The technique relies on minimizing the sum of squared differences between actual and predicted Y values, a method known as least squares.


# **2)  What are the key assumptions of Simple Linear Regression?**

**Ans)-** The key assumptions of **Simple Linear Regression** ensure the validity and reliability of the model’s results. These assumptions are:

1. **Linearity**:
   There is a **linear relationship** between the independent variable (X) and the dependent variable (Y). This means the effect of X on Y is constant across all values of X.

2. **Independence of Errors**:
   The residuals (differences between observed and predicted values) are **independent** of each other. This means the error terms are not correlated, which is especially important in time series data.

3. **Homoscedasticity**:
   The residuals have **constant variance** across all levels of X. In other words, the spread of errors should be roughly the same regardless of the value of X.

4. **Normality of Errors**:
   The residuals are **normally distributed**. This assumption is particularly important when making inferences (e.g., hypothesis testing or confidence intervals).

5. **No Multicollinearity (in multiple regression)**:
   Though this applies to multiple regression, it’s worth noting that in simple linear regression (with only one predictor), this assumption is naturally satisfied.

Violating these assumptions can lead to biased or inefficient estimates and unreliable conclusions. Checking diagnostic plots (like residual plots or Q-Q plots) can help assess whether these assumptions hold.




# **3) What is heteroscedasticity, and why is it important to address in regression models? **


**Ans)-** **Heteroscedasticity** is a condition in regression analysis where the variance of the residuals (errors) is not constant across all levels of the independent variable. This violates one of the key assumptions of linear regression, which assumes homoscedasticity, meaning the residuals should have equal variance regardless of the value of the predictor variable. When heteroscedasticity is present, it can lead to several problems. Although the estimated regression coefficients may still be unbiased, the standard errors associated with them become unreliable. This can result in incorrect conclusions about the significance of predictors, as p-values and confidence intervals may be distorted. Heteroscedasticity can also reduce the efficiency of the model, making predictions less reliable, especially in regions where the variance of errors is high. Detecting heteroscedasticity can be done visually through residual plots or statistically using tests like the Breusch-Pagan or White test. If detected, it can be addressed by transforming the dependent variable (such as applying a log or square root), using weighted least squares, or applying robust standard errors. Addressing heteroscedasticity is important to ensure the validity of the regression results and the reliability of any conclusions drawn from the model.



# **4) What is Multiple Linear Regression?**

**Ans)-** **Multiple Linear Regression** is a statistical technique used to model the relationship between **one dependent variable** and **two or more independent variables**. It is an extension of simple linear regression, which involves only one predictor. The goal of multiple linear regression is to understand how multiple factors simultaneously influence the outcome and to predict the value of the dependent variable based on the values of the independent variables.

The general form of the multiple linear regression equation is:

$$
Y = \beta_0 + \beta_1 X_1 + \beta_2 X_2 + \dots + \beta_n X_n + \varepsilon
$$

Here, $Y$ is the dependent variable, $X_1, X_2, \dots, X_n$ are the independent variables, $\beta_0$ is the intercept, $\beta_1, \beta_2, \dots, \beta_n$ are the coefficients for each predictor, and $\varepsilon$ is the error term. Each coefficient represents the change in the dependent variable for a one-unit change in the corresponding independent variable, assuming all other variables are held constant.

Multiple linear regression is widely used in fields such as economics, engineering, social sciences, and medicine to analyze complex relationships and make predictions. However, it requires that certain assumptions—such as linearity, independence of errors, homoscedasticity, and no multicollinearity—are met for the model to be valid and reliable.


# ** 5) What is polynomial regression, and how does it differ from linear regression? **

**Ans)-** **Polynomial regression** is a type of regression analysis used to model the relationship between a dependent variable and an independent variable when that relationship is **nonlinear**. Unlike simple linear regression, which fits a straight line to the data, polynomial regression fits a **curved line** by introducing higher-degree terms of the independent variable (such as $X^2, X^3,$ etc.) into the model.

The general form of a polynomial regression equation is:

$$
Y = \beta_0 + \beta_1 X + \beta_2 X^2 + \beta_3 X^3 + \dots + \beta_n X^n + \varepsilon
$$

Here, $Y$ is the dependent variable, $X$ is the independent variable, $\beta_0, \beta_1, ..., \beta_n$ are the coefficients, and $\varepsilon$ is the error term. The key difference from linear regression is that polynomial regression can capture **curvature** in the data, making it more flexible for fitting complex patterns.

While polynomial regression is still considered a form of **linear regression** (because it's linear in the coefficients), it differs from simple linear regression in its ability to model **nonlinear relationships**. However, increasing the polynomial degree too much can lead to **overfitting**, where the model fits the training data very well but performs poorly on unseen data. Therefore, it's important to balance flexibility with simplicity when choosing the degree of the polynomial.


```python
#ques 6- Implement a Python program to fit a Simple Linear Regression model to the following sample data: ● X = [1, 2, 3, 4, 5] ● Y = [2.1, 4.3, 6.1, 7.9, 10.2]. Plot the regression line over the data points



import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # Reshape to 2D array for sklearn
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Get model parameters
slope = model.coef_[0]
intercept = model.intercept_
print(f"Fitted line: Y = {intercept:.2f} + {slope:.2f}X")

# Predict Y values
Y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, Y, color='blue', label='Actual Data')
plt.plot(X, Y_pred, color='red', label='Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#ques 7 Fit a Multiple Linear Regression model on this sample data: ● Area = [1200, 1500, 1800, 2000] ● Rooms = [2, 3, 3, 4] ● Price = [250000, 300000, 320000, 370000] Check for multicollinearity using VIF and report the results.



import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from statsmodels.stats.outliers_influence import variance_inflation_factor
import statsmodels.api as sm

# Sample data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Independent variables (X) and dependent variable (Y)
X = df[['Area', 'Rooms']]
Y = df['Price']

# Fit the multiple linear regression model
model = LinearRegression()
model.fit(X, Y)

# Print the regression equation
print("Regression Equation:")
print(f"Price = {model.intercept_:.2f} + {model.coef_[0]:.2f}*Area + {model.coef_[1]:.2f}*Rooms\n")

# Add constant for VIF calculation
X_with_const = sm.add_constant(X)

# Calculate VIF
vif_data = pd.DataFrame()
vif_data['Feature'] = X_with_const.columns
vif_data['VIF'] = [variance_inflation_factor(X_with_const.values, i)
                   for i in range(X_with_const.shape[1])]

print("Variance Inflation Factor (VIF):")
print(vif_data)
```

```python
# ques 8- Implement polynomial regression on the following data: ● X = [1, 2, 3, 4, 5] , ● Y = [2.2, 4.8, 7.5, 11.2, 14.7] Fit a 2nd-degree polynomial and plot the resulting curve.



import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Transform features to polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Fit the model
model = LinearRegression()
model.fit(X_poly, Y)

# Generate predictions (smooth curve)
X_curve = np.linspace(1, 5, 100).reshape(-1, 1)
X_curve_poly = poly.transform(X_curve)
Y_curve = model.predict(X_curve_poly)

# Plotting
plt.scatter(X, Y, color='blue', label='Data Points')
plt.plot(X_curve, Y_curve, color='red', label='2nd Degree Polynomial Fit')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()

# Print the coefficients
print(f"Intercept: {model.intercept_:.2f}")
print(f"Coefficients: {model.coef_}")
```

```python
# ques 9-  Create a residuals plot for a regression model trained on this data: ● X = [10, 20, 30, 40, 50] ● Y = [15, 35, 40, 50, 65] Assess heteroscedasticity by examining the spread of residuals.



import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y and calculate residuals
Y_pred = model.predict(X)
residuals = Y - Y_pred

# Plot residuals vs predicted values
plt.scatter(Y_pred, residuals, color='purple')
plt.axhline(y=0, color='black', linestyle='--')
plt.xlabel('Predicted Values')
plt.ylabel('Residuals')
plt.title('Residuals Plot')
plt.grid(True)
plt.show()
```

# **10)  Imagine you are a data scientist working for a real estate company. You need to predict house prices using features like area, number of rooms, and location. However, you detect heteroscedasticity and multicollinearity in your regression model. Explain the steps you would take to address these issues and ensure a robust model. **


**Ans)-**  As a data scientist facing **heteroscedasticity** and **multicollinearity** in a house price prediction model using features like area, number of rooms, and location, I would take the following steps to address these issues and build a robust model:

---

### 1. **Addressing Heteroscedasticity:**

* **Transform the dependent variable:**
  Applying a transformation like the logarithm (`log(price)`) often stabilizes the variance of residuals. Other transformations such as square root or Box-Cox can also help.

* **Use weighted least squares (WLS):**
  If the variance of errors depends on certain features, WLS assigns weights inversely proportional to the variance, improving model reliability.

* **Robust standard errors:**
  If transformations aren’t ideal, using heteroscedasticity-robust standard errors can produce valid inference without changing the model structure.

* **Examine and incorporate missing variables:**
  Sometimes heteroscedasticity arises from omitted variables related to variance. Ensuring the model includes key predictors can reduce it.

---

### 2. **Addressing Multicollinearity:**

* **Check correlation and VIF scores:**
  Identify highly correlated predictors or those with high Variance Inflation Factor (VIF).

* **Remove or combine correlated variables:**
  For example, if area and number of rooms are highly correlated, consider dropping one or combining them into a new feature like “average room size.”

* **Feature engineering:**
  Create meaningful composite variables or apply dimensionality reduction techniques like **Principal Component Analysis (PCA)** to reduce collinearity.

* **Regularization techniques:**
  Methods like **Ridge Regression** or **Lasso** add penalties to coefficient sizes and can reduce the impact of multicollinearity while improving model generalization.

---

### 3. **General Best Practices:**

* **Validate model assumptions:**
  Continuously check residual plots and diagnostics after adjustments.

* **Cross-validation:**
  Use cross-validation to ensure the model performs well on unseen data.

* **Incorporate domain knowledge:**
  Work with real estate experts to understand which features are most meaningful and how to interpret transformations or engineered variables.

---

By combining these steps, the model will better satisfy regression assumptions, reduce biases, and yield more accurate and reliable house price predictions.
